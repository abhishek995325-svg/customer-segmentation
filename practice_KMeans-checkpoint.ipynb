#### Elbow method

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import KMeans
```

```python
# wcss = within cluster sum of squares
wcss = []
for i in range(1,21):
    model = KMeans(n_clusters=i)
    model.fit(x)
    wcss.append(model.inertia_)

print('wcss')
print(wcss)
```

```python
plt.plot(list(range(1,21)) ,wcss)
plt.xlabel('Num of clusters')
plt.ylabel('WCSS')
plt.title('Elbow method')
plt.xticks(list(range(1,21)))
plt.grid()
plt.show()
```

```python
k5 = KMeans(n_clusters=5)
k5.fit(x)
```

```python
ypred_k5 = k5.predict(x)
print(len(ypred_k5))
```

```python
df['Cls5'] = ypred_k5
df['Cls5'].value_counts()
```

```python
cent5 = k5.cluster_centers_
print(cent5)
```

```python
sns.scatterplot(data = df , x='AnnInc' , y = 'SScore' , hue= 'Cls5' , palette='Set1')
plt.scatter(cent5[:,0] , cent5[:,1],label='Centroids' ,marker='*' , s =120 , color='black')
plt.legend()
plt.show()
```
